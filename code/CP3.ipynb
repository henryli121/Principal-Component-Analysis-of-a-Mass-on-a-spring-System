```python
import numpy as np
import os
import matplotlib.pyplot as plt
import matplotlib.image as image
import scipy.io
import copy

from numpy import linalg as LA
```

```python
cwd = os.getcwd()
Xt1_1= scipy.io.loadmat('Xt1_1.mat')['Xt1_1']
Xt2_1= scipy.io.loadmat('Xt2_1.mat')['Xt2_1']
Xt3_1= scipy.io.loadmat('Xt3_1.mat')['Xt3_1']
```

```python
print(Xt1_1.shape, Xt2_1.shape,Xt3_1.shape)
```

```python
length_1 = 226
Xt2_1 = Xt2_1[:,:length_1]
Xt3_1 = Xt3_1[:,:length_1]
```

```python
Xt1_1 = (Xt1_1.T - np.mean(Xt1_1,axis = 1)).T
Xt2_1 = (Xt2_1.T - np.mean(Xt2_1,axis = 1)).T
Xt3_1 = (Xt3_1.T - np.mean(Xt3_1,axis = 1)).T
```

```python
mat_1 = np.vstack((Xt1_1,Xt2_1,Xt3_1))
u_1, s_1, vh_1 = LA.svd(mat_1, full_matrices=False)
#lamda_1  =np.diag(s_1)**2
sig_1  =np.diag(s_1)
Y_1 = u_1.T@mat_1
A1 = copy.copy(Y_1)
```

```python
energy_1 = []
for i in range(6):
    E = np.sum(sig_1[0:i+1]**2)/np.sum(sig_1**2)
    energy_1 = np.append(energy_1,E)

A2 = copy.copy(energy_1.reshape((6,1)))
```

```python
energy_1
```

```python
cum_energy_1 = np.cumsum(energy_1)

fig = plt.figure(figsize = (6,4))
ax = plt.gca()
ax.scatter(range(6),np.log(cum_energy_1),facecolors='none',edgecolors='k')
ax.grid(axis='y',linewidth=0.5)
plt.title('Cumulative Energy for Ideal Case')
plt.xlabel('x')
plt.ylabel('Cumulative energy in log scale')
#plt.savefig('energy_1.png',dpi=600)
plt.show()
```

```python
n = 2
U_1 = u_1
V_1 = vh_1.T
prox_mat_1 = U_1[:,0:n]@sig_1[0:n,0:n]@V_1[:,0:n].T
A3 = copy.copy(prox_mat_1)
```

```python
fig = plt.figure(figsize = (8,6))
for i in range(6):
    plt.plot(prox_mat_1[i],label = i+1,linewidth = 1)

plt.grid(axis='both',linewidth=0.5)
plt.legend()
plt.title('SVD Approximation for Ideal Case')
plt.xlabel('time t')
plt.ylabel('amplitude')
#plt.savefig('approx_1.png',dpi=600)
plt.show()
```
